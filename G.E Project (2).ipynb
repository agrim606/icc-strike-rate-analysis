```python
def str_to_int(x:str):
    if x[-1].isdigit():
        return int(x)
    elif len(x)==0:
        return 0
    else:
        x=x[0:-1]
        if len(x) == 0:
            return 0
        else:
            return int(x)
df["H.S"]=df["H.S"].apply(str_to_int)
```

# Data Analysis on ICC tournament teams

```python
list_Country= ["(INDIA)","(PAK)","(AUS)","(ENG)","(SA)","(NZ)","(WI)","(AFG)","(SL)","(BDESH)"]
Average_SR=[]
for i in list_Country:
    Average_SR.append(df[df["Country"]==i]["SR"].mean())
    
```

```python
Average_SR
```

```python
df[df["Country"]=="(INDIA)"].sum()
```

```python
df[df["Country"]=="(INDIA)"].count()
```

```python
sr=pd.Series(Average_SR,index=list_Country)
```

```python
sr
```

```python
plt.bar(sr.index,sr.values,color="g")
plt.title("Strike Rate of ICC Tournament Teams")
```

```python
plt.barh(t["Player"],t["SR"],color=Col1,alpha=0.5)
plt.title("Strike Rate Of Top 10 Highest 4s Hitters")
plt.grid(True)
plt.ylabel("Players Name")
plt.xlabel("Strike rate")
```

```python
for i in range(10):
    plt.plot(["Average","6s","SR"],t.loc[i,["Average","6s","SR"]],label=t.loc[i,"Player"])
plt.legend()
plt.grid(True)
plt.title("Comparsion between top 10 4s hitters")
```
